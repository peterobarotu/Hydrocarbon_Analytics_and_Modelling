```python
# Improved function with flexible number of columns in subplots
import math
def plot_dist(data, ncols=2):  # ncols is the number of subplots per row
    num_plots = len(data.columns)  # Number of columns in the DataFrame
    nrows = math.ceil(num_plots / ncols)  # Calculate number of rows required

    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(4*ncols, 3.3*nrows))  # Adjust figure size dynamically

    # Flatten axes if there are multiple rows/columns of subplots
    axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]

    # Loop through each column in the DataFrame and create a histogram plot for each
    for i, col in enumerate(data.columns):
        sns.histplot(data[col], kde=True, ax=axes[i])  # Use the i-th subplot
        axes[i].set_title(f'Distribution of {col}', fontsize=8)
        axes[i].set_xlabel(f'{col}', fontsize=8)
        axes[i].set_ylabel('Frequency', fontsize=9)

    # Remove any extra axes (if num_plots < nrows*ncols)
    for j in range(num_plots, nrows*ncols):
        fig.delaxes(axes[j])  # Delete the unused axes
    
    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()
```

```python
inject_vol = west_net.groupby('GAS SUPPLIER')['INJECTED VOL. MMSCF'].mean()
inject_press = west_net.groupby('GAS SUPPLIER')['INJ. PRESS. BARG'].mean()
inject_data = pd.concat([inject_vol, inject_press], axis=1)

plot_dist(inject_data)
```

```python
suppliers_vol = inject_data.sort_values('INJECTED VOL. MMSCF', ascending=False).reset_index()
suppliers_vol['% VOLUME'] = suppliers_vol['INJECTED VOL. MMSCF']/ suppliers_vol['INJECTED VOL. MMSCF'].sum() * 100
suppliers_vol
```

```python

# Sort by 'INJECTED VOL. MMSCF' in descending order to apply Pareto principle
suppliers_vol = suppliers_vol.sort_values('INJECTED VOL. MMSCF', ascending=False)

# Calculate cumulative percentage for Pareto line
suppliers_vol['Cumulative Percentage'] = suppliers_vol['INJECTED VOL. MMSCF'].cumsum() / suppliers_vol['INJECTED VOL. MMSCF'].sum() * 100

# Create the figure and axis objects
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plotting the bar plot for INJECTED VOL. MMSCF on the primary y-axis
ax1.bar(suppliers_vol['GAS SUPPLIER'], suppliers_vol['INJECTED VOL. MMSCF'], color='blue', alpha=0.6, label='Injected Vol. (MMSCF)')
ax1.set_xlabel('Gas Supplier')
ax1.set_ylabel('Injected Volume (MMSCF)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create the secondary y-axis and plot the cumulative percentage for the Pareto line
ax2 = ax1.twinx()
ax2.plot(suppliers_vol['GAS SUPPLIER'], suppliers_vol['Cumulative Percentage'], color='green', marker='o', label='Cumulative %', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylim([0, 110])  # Set limit slightly above 100 to ensure line visibility

# Adding labels, title, and legends
plt.title('Injected Volume and Cumulative Percentage (Pareto) by Gas Supplier for 01-03-2024')
fig.tight_layout()  # Adjust layout for better appearance

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_injected_volume_and_pressure(df, volume_col, pressure_col):
    """
    Plot injected volume and pressure on a twin y-axis plot.
    
    :param df: DataFrame containing the data
    :param volume_col: The column name for injected volume
    :param pressure_col: The column name for pressure
    """
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Plot injected volume on the left y-axis
    ax1.set_xlabel('GAS SUPPLIER')
    ax1.set_ylabel(f'{volume_col} (Injected Volume)', color='tab:blue')
    ax1.plot(df.index, df[volume_col], color='tab:blue', label=volume_col)
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    plt.xticks(rotation=45)

    # Create a twin y-axis to plot pressure
    ax2 = ax1.twinx()
    ax2.set_ylabel(f'{pressure_col} (Pressure)', color='tab:red')
    ax2.plot(df.index, df[pressure_col], color='tab:red', label=pressure_col)
    ax2.tick_params(axis='y', labelcolor='tab:red')

    

    # Add a title and adjust layout
    plt.title(f'{volume_col} vs {pressure_col}')
    fig.tight_layout()  # Adjust to prevent overlap
    plt.show()
```

```python
def supplier_summary(df, cols):
    all_supplier = []
    for supplier in df['GAS SUPPLIER'].unique():
        supplier_df = df[df['GAS SUPPLIER'] == supplier]
        agg_df = supplier_df[cols].agg(['min', 'max', 'mean', 'sum']).reset_index(names='Aggregate')
        agg_df = pd.concat([pd.Series([supplier]*4, name='GAS SUPPLIER'), agg_df], axis=1)
        all_supplier.append(agg_df)
    return pd.concat(all_supplier).round(2)    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sort the DataFrame by 'INDUSTRY' to ensure grouping
offtaker_vol_perf = offtaker_vol_perf.sort_values(by='INDUSTRY')

# Remove rows with 0 allocated volume values
offtaker_vol_perf = offtaker_vol_perf[offtaker_vol_perf['ALLOCATED VOL. MMSCF'] > 0]

# Group by 'INDUSTRY'
grouped = offtaker_vol_perf.groupby('INDUSTRY')

# Plotting
fig, ax = plt.subplots(figsize=(12, 7))

# Define the number of industries and the number of unique shipper/offtaker entries
n_groups = len(offtaker_vol_perf['SHIPPER(S)/OFFTAKER(S)'].unique())
n_industries = len(offtaker_vol_perf['INDUSTRY'].unique())

# Define positions for each group (i.e., industry)
bar_width = 0.4  # Width of each bar
index = np.arange(len(offtaker_vol_perf))

# Initialize the offsets for each industry group
industry_colors = {'LDC': 'green', 'IPP': 'blue', 'others': 'orange'}
offset = 0

# Plot each industry group
for i, (industry, group) in enumerate(grouped):
    bar_positions = index[offset:offset + len(group)]  # Positions for each industry bar
    ax.bar(bar_positions, group['ALLOCATED VOL. MMSCF'], bar_width, label=industry, color=industry_colors[industry])
    offset += len(group)

# Labeling
ax.set_xlabel('Shipper/Offtaker')
ax.set_ylabel('Allocated Volume (MMSCF)')
ax.set_title('Allocated Volume by Shipper/Offtaker and Industry')

# Set xticks and labels
ax.set_xticks(index)
ax.set_xticklabels(offtaker_vol_perf['SHIPPER(S)/OFFTAKER(S)'], rotation=70)

# Add legend in upper right corner
ax.legend(title='Industry', loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
def get_offtaker_perf(groupby_obj, offtaker):
    offtaker_df =  (groupby_obj.get_group(offtaker)[other_cols]
                    .agg(['min', 'max', 'mean', 'sum'])
                    .reset_index(names='Aggregate')
                   )
    offtaker_df = pd.concat([pd.Series([offtaker]*4, name='SHIPPER(S)/OFFTAKER(S)'), offtaker_df], axis=1)
    return offtaker_df
```
